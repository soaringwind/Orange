```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('./hw1/train.csv', encoding='big5')
```

```python
df.drop('observation', axis=1, inplace=True)
```

```python
df[df=='NR'] = 0
```

```python
raw_data = np.array(df)
```

```python
raw_data.shape
```

```python
month_data = {}
for month in range(12):
    sample = np.empty([18,480])
    for day in range(20):
        sample[:, day*24:(day+1)*24] = raw_data[18*(20*month+day):18*(20*month+day+1), :]
    month_data[month] = sample
```

```python
month_data
```

```python
train_data = np.empty([12*471, 18*9], dtype=float)
train_label = np.empty([12*471, 1], dtype=float)
for month in range(12):
    for day in range(20):
        for hour in range(24):
            if day == 19 and hour > 14:
                continue
            train_data[month*471+day*24+hour, :] = month_data[month][:, day*24+hour:day*24+hour+9].reshape(1,-1)
            train_label[month*471+day*24+hour, 0] = month_data[month][9, day*24+hour+9]
```

```python
dim = 18*9
w = np.zeros([dim, 1])
# train_data = np.concatenate((np.ones([len(train_data), 1]), train_data), axis=1).astype('float')
lr = 0.0000000001
iter_time = 300
adagrad = np.zeros([dim, 1])
eps = 0.000001
for i in range(iter_time):
    loss = np.sum(np.power((np.dot(train_data, w)-train_label), 2))
    if i%10 == 0:
        print('迭代的次数：%i，损失值：%f'%(i, loss))
        # gradient = (np.dot(train_data.transpose(), np.dot(train_data, w)-train_label))/(loss*(len(train_data)))
        gradient = (-2.0) * np.dot(train_data.transpose(),train_label-np.dot(train_data, w))
        # adagrad += gradient**2
        # w -= lr * (gradient/np.sqrt(adagrad+eps))
        w -= lr * gradient
```
